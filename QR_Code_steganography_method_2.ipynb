```python
import qrcode
from cryptography.fernet import Fernet
import base64
import os

# Generate and save an encryption key (Run this once)
def generate_key():
    key = Fernet.generate_key()
    with open("secret.key", "wb") as key_file:
        key_file.write(key)

# Load the encryption key
def load_key():
    if not os.path.exists("secret.key"):
        generate_key()
    with open("secret.key", "rb") as key_file:
        return key_file.read()

# Encrypt the hidden message
def encrypt_message(secret_message):
    key = load_key()
    cipher = Fernet(key)
    encrypted_msg = cipher.encrypt(secret_message.encode())
    return base64.b64encode(encrypted_msg).decode()  # Encode to make it QR-friendly

# Create a QR code with a website link and encrypted hidden message
def create_encrypted_qr():
    website_url = input("Enter the website URL: ").strip()
    hidden_msg = input("Enter the hidden message: ").strip()

    encrypted_hidden = encrypt_message(hidden_msg)

    # Store encrypted message inside the URL
    full_data = f"{website_url}#hidden={encrypted_hidden}"

    qr = qrcode.QRCode(error_correction=qrcode.constants.ERROR_CORRECT_H)
    qr.add_data(full_data)
    qr.make()
    img = qr.make_image(fill="black", back_color="white")

    qr_path = "secure_qr.png"
    img.save(qr_path)

    print(f"‚úÖ QR Code saved as {qr_path}")
    print(f"üîó Scan it to visit: {website_url}")
    print("üîí The hidden message is encrypted inside the QR.")

# Run QR Code Creation
create_encrypted_qr()
```

```python
from pyzbar.pyzbar import decode
from PIL import Image
import base64

# Decrypt hidden message
def decrypt_message(encrypted_message):
    key = load_key()
    cipher = Fernet(key)
    decrypted_msg = cipher.decrypt(base64.b64decode(encrypted_message)).decode()
    return decrypted_msg

# Extract hidden message from QR
def extract_encrypted_qr():
    qr_image_path = input("Enter the path of the QR code image: ").strip()

    try:
        qr_img = Image.open(qr_image_path)
        decoded_data = decode(qr_img)

        if not decoded_data:
            print("‚ùå No QR code found in the image.")
            return

        decoded_text = decoded_data[0].data.decode()

        # Extract hidden part from the URL
        if "#hidden=" in decoded_text:
            website_link, hidden_encoded = decoded_text.split("#hidden=", 1)
            try:
                hidden_message = decrypt_message(hidden_encoded)
                print(f"\nüîó Website: {website_link}")
                print(f"üîí Hidden Message: {hidden_message}")
            except Exception:
                print("‚ùå Decryption Failed: Invalid Key!")
        else:
            print(f"\nüîó Website: {decoded_text}")
            print("‚ùå No hidden data found.")

    except FileNotFoundError:
        print("‚ùå File not found. Please enter a valid QR code image path.")

# Run QR Code Extraction
extract_encrypted_qr()
```
